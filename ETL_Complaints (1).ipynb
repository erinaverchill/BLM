```python
import pandas as pd
```

```python
#import complaints excel
complaints_df = pd.read_csv("CCAPO_2016-2018_Agency_0.csv")
complaints_df
```

```python
complaints_df.columns
```

```python
#edit/modify columns with total complaints only
total_complaints_df = pd.DataFrame(complaints_df[{'NCIC','County',
                                                  'AgencyName',
                                                  'ComplaintYear', 
                                                  'TotalComplaintsReported',
                                                  'TotalComplaintsSustained',
                                                  'TotalComplaintsExonerated',
                                                  'TotalComplaintsNotSustained',
                                                  'TotalComplaintsUnfounded',
                                                  'TotalComplaintsPending'}])

total_complaints_df = total_complaints_df.reindex(columns = ['NCIC','County',
                                        'AgencyName',
                                        'ComplaintYear', 
                                        'TotalComplaintsReported',
                                        'TotalComplaintsSustained',
                                        'TotalComplaintsExonerated',
                                        'TotalComplaintsNotSustained',
                                        'TotalComplaintsUnfounded',
                                        'TotalComplaintsPending'])

total_complaints_df = total_complaints_df.rename(columns = {
                                        'AgencyName':'Agency',
                                        'ComplaintYear': 'Complaint Year', 
                                        'TotalComplaintsReported': 'Complaints Reported',
                                        'TotalComplaintsSustained':'Complaints Sustained',
                                        'TotalComplaintsExonerated': 'Complaints Exonerated',
                                        'TotalComplaintsNotSustained':'Complaints Not Sustained',
                                        'TotalComplaintsUnfounded': 'Complaints Unfounded',
                                        'TotalComplaintsPending': 'Complaints Pending'})

total_complaints_df

```

```python
#find sum of complaints for all rows as total complaints
total =0
for index, row in total_complaints_df.iterrows():
    total = row[4] + row[5] + row[6] + row[7] + row[8] + row[9]
    total_complaints_df.loc[index,'Total Complaints']= total

total_complaints_df.head()
```

```python
complaints_per_county_df = total_complaints_df.groupby(['County', 'Complaint Year']).sum()
complaints_per_county_df
```

```python
#find percentages
complaints_per_county_df['Percentage Sustained'] = 100 * complaints_per_county_df['Complaints Sustained']/complaints_per_county_df['Complaints Reported']
complaints_per_county_df['Percentage Exonerated'] = 100 * complaints_per_county_df['Complaints Exonerated']/complaints_per_county_df['Complaints Reported']
complaints_per_county_df['Percentage Not Sustained'] = 100 * complaints_per_county_df['Complaints Not Sustained']/complaints_per_county_df['Complaints Reported']
complaints_per_county_df['Percentage Unfounded'] = 100 * complaints_per_county_df['Complaints Unfounded']/complaints_per_county_df['Complaints Reported']
complaints_per_county_df['Percentage Pending'] = 100 * complaints_per_county_df['Complaints Pending']/complaints_per_county_df['Complaints Reported']
  
    
#drop rows with null values
complaints_per_county_df.dropna(inplace = True)
complaints_per_county_df

```

```python
# to csv
complaints_per_county_df.to_csv('police_complaints_per_county.csv')
```
